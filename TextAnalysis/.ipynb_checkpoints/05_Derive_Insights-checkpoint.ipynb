## Step 5. 인사이트 도출

####  최종적으로 추출된 토큰들을 가지고 인사이트를 도출해 봅시다.

```python
# -*- coding: utf-8 -*-
import pandas as pd
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### 1. Word Count - Table

```python
data = pd.read_excel('./Data/tokens_final.xlsx')
```

```python
data.head()
```

#### 전체 단어 corpus에서 각각의 Frequency를 측정한다.

```python
total_corpus = []
for index, row in data.iterrows():
    corpus = row['Tokens'].split('/')
    total_corpus.extend(corpus)
```

```python
counts = Counter(total_corpus)
freq = counts.most_common(100)
```

```python
pd.DataFrame(freq, columns=['단어','빈도수'])
```

### 2. Word Count - Histogram

#### matplotlib를 이용하여 histogram그리기

```python
words = [w[0] for w in freq]
counts = [w[1] for w in freq]
```

```python
indices = np.arange(len(words))
width = 0.7
fig = plt.figure(figsize=(80,40))
plt.bar(indices, counts, width)
plt.xticks(indices + width * 0.5, words)
plt.show()
```

#### 3. Word Cloud

#### Python package를 이용하여 Word cloud를 그립니다. 

```python
from scipy.misc import imread
from wordcloud import WordCloud
```

```python
def grey_color(word, font_size, position, orientation, random_state=None, **kwargs):
    # return 'hsl(0, 0%%, %d%%)' % random.randint(50, 100)
    return 'hsl(0, 0%%, %d%%)' % 30
```

```python
def draw_word_cloud_freq(freq, cluster_name, color_func, save_path, random_state = 42, show=False):
    """
    word_cloud로 그려줌
    """
    circle_mask = imread("./circle.png") # 모서리가 둥근 네모 이미지를 설정
    wc = WordCloud(background_color="white", 
                   font_path = './NanumBarunGothic.ttf', #폰트 설정
                   width=500,height=300,
                   max_words=200,
                   max_font_size=150,
                   min_font_size=15,
                   mask=circle_mask, 
                   prefer_horizontal=1, random_state = 0)
    wc.generate_from_frequencies(freq)
    wc.recolor(color_func=color_func, random_state=random_state)
    plt.figure(figsize=(8,6))
    plt.imshow(wc, cmap=plt.cm.Pastel1)
    plt.title(cluster_name)
    plt.axis("off")
    plt.savefig(save_path + cluster_name + ".png")
    if show == True:
        plt.show()
```

```python
%matplotlib inline
save_path = './'
```

```python
draw_word_cloud_freq(freq, 'NAVER_Finance_2017_08_30', grey_color, save_path)
```
